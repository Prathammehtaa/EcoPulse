```python
import pandas as pd

def ingest_eco_pulse_history(file_path):
    # 1. Load data skipping the 3 Open-Meteo metadata rows
    df = pd.read_csv(file_path, skiprows=3)
    
    # 2. Convert 'time' to UTC datetime objects
    df['time'] = pd.to_datetime(df['time'])
    
    # 3. Rename columns for Machine Learning (removing units/special characters)
    column_mapping = {
        'time': 'timestamp_utc',
        'temperature_2m (°C)': 'temp_2m_c',
        'relative_humidity_2m (%)': 'humidity_pct',
        'cloud_cover (%)': 'cloud_cover_pct',
        'rain (mm)': 'rain_mm',
        'snowfall (cm)': 'snowfall_cm',
        'wind_speed_100m (km/h)': 'wind_speed_100m_kmh'
    }
    df = df.rename(columns=column_mapping)
    
    # 4. Confirm it starts from January 1st, 2025
    df = df[df['timestamp_utc'] >= '2025-01-01'].copy()
    
    print(f"Ingested {len(df)} hourly records.")
    print(f"Range: {df['timestamp_utc'].min()} to {df['timestamp_utc'].max()}")
    return df

# Execute Ingestion
weather_df = ingest_eco_pulse_history('open-meteo-history.csv')
weather_df.to_csv('eco_pulse_weather_2025_cleaned.csv', index=False)
```
